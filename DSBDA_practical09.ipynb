# Visualization

```python
def fun1(value):
    if (value == "male"):
        return 1
    else:
        return 0
```

```python
def fun2(value):
    if (value == 'S'):
        return 0
    elif (value == 'C'):
        return 1
    elif (value == 'Q'):
        return 2
    else:
        return 0
```

```python
df["Sex"] = df["Sex"].apply(fun1)
```

```python
df["Embarked"] = df["Embarked"].apply(fun2)
```

```python
df = df.drop("Cabin", axis=1)
```

```python
df.shape
```

```python
px.box(df["Sex"], df["Age"], color=df["Survived"])
```

```python
plt.figure(figsize=(10,7))
box = sns.boxplot(df["Sex"], df["Age"], hue=df["Survived"])
plt.show()
```
